```python
# Import the packages we will use
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
loop_mean = np.load('arrs/densities/mean_density_loops_mean_rao.npy')
loop_sum = np.load('arrs/densities/mean_density_loops_sum_rao.npy')
btw_mean = np.load('arrs/densities/mean_density_btw_mean_rao.npy')
btw_sum = np.load('arrs/densities/mean_density_btw_sum_rao.npy')
```

```python
def plot_loop_and_btw_density(loop, btw, isLog=True):
    fig = plt.figure()
    ax = plt.axes()

    if isLog:
        plt.plot(np.log(range(loop.shape[0])), np.log(loop), label='loop');
        plt.plot(np.log(range(btw.shape[0])), np.log(btw), label='btw');
    
    else:
        plt.plot(range(loop.shape[0]), loop, label='loop');
        plt.plot(range(btw.shape[0]), btw, label='btw');
        
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
```

```python
plot_loop_and_btw_density(loop_mean, btw_mean)
```

```python
plot_loop_and_btw_density(loop_mean, btw_mean, isLog=False)
```

```python
plot_loop_and_btw_density(loop_sum, btw_sum)
```

```python
plot_loop_and_btw_density(loop_sum, btw_sum, isLog=False)
```
