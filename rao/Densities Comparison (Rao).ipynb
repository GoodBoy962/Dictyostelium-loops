```python
# Import the packages we will use
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
loop_mean = np.load('arrs/densities/avg_density_loops_mean_rao.npy')
loop_sum = np.load('arrs/densities/avg_density_loops_sum_rao.npy')
interloops_mean = np.load('arrs/densities/avg_density_interloops_mean_rao.npy')
interloops_sum = np.load('arrs/densities/avg_density_interloops_sum_rao.npy')
```

```python
def plot_loop_and_interloops_density(loop, interloops, isLog=True):
    fig = plt.figure()
    ax = plt.axes()

    if isLog:
        plt.plot(np.log(range(loop.shape[0])), np.log(loop), label='loop');
        plt.plot(np.log(range(interloops.shape[0])), np.log(interloops), label='interloops');
    
    else:
        plt.plot(range(loop.shape[0]), loop, label='loop');
        plt.plot(range(interloops.shape[0]), interloops, label='interloops');
        
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
```

```python
plot_loop_and_interloops_density(loop_mean, interloops_mean)
```

```python
plot_loop_and_interloops_density(loop_mean, interloops_mean, isLog=False)
```

```python
plot_loop_and_interloops_density(loop_sum, interloops_sum)
```

```python
plot_loop_and_interloops_density(loop_sum, interloops_sum, isLog=False)
```
