```python
resolution = 10 ** 4
```

###### Read loops coordinates, sizes and etc

```python
loops_info = pandas.read_csv('GSE63525_GM12878_primary+replicate_HiCCUPS_looplist.txt', sep='\t',
                             dtype = {'x1': np.long, 'x2': np.long, 'y1': np.long, 'y2': np.long})
loops_info[['x1', 'x2', 'y1', 'y2', 'x', 'y']] = loops_info[['x1', 'x2', 'y1', 'y2', 'centroid1', 'centroid2']].apply(
    lambda x : (x / resolution).astype('int64')
)
```

```python
def plot_resized_average_with_window(chr, loop_new_size, is_loop_window):
    loops_info_chr = loops_info[loops_info.chr1 == str(chr)]
    container = LoopChromosomeContainer(np.load('arrs/10kb_chr' + str(chr) + '_norm.npy'), loops_info_chr)
    name = 'average_loops/rao_resized_norm_average_loop_chr' + str(chr)
    if is_loop_window:
        name = 'average_loops/rao_resized_with_window_norm_average_loop_chr' + str(chr)
    loop_avg = container.avg_loop_resized(loop_new_size, is_loop_window)
    loops_utils.plot_HiC(loops_utils.z_norm_zero_middle(loop_avg), 
                         figsize=(10, 10),
                         name=name, 
                         is_loop_window=is_loop_window)
    return container, loop_avg
```
